```python
scraped=[]

base_url='https://nigeriapropertycentre.com/for-rent/flats-apartments/lagos/showtype?page='

items = int(soup.find_all('span',{'class':'pagination-results'})[0].get_text().strip().split()[-1].replace(',',''))
listings = 30
page_nr = int(items/listings)

for page in range(1,int(page_nr),1):
    print(base_url+str(page))
    r= requests.get(base_url+str(page))
    c= r.content
    soup= BeautifulSoup(c,"html.parser")
        
    classes = ["itemListElement"]
    for class_ in classes:
        real=soup.find_all("div",{"itemprop":class_})

        for i in list(range(0,len(real))):
            d={}
            d["page" ] = page
            try:
                d["title"]=real[i].find('h3', {'itemprop':'name'}).get_text().strip()
            except (IndexError,TypeError,AttributeError) as e:
                d["title"]= None
            try:
                d["description"]=real[i].find('h4', {'class':'content-title'}).get_text().strip()
            except (IndexError,TypeError,AttributeError) as e:
                d["description"]= None
            try:
                d["bedrooms"]= re.findall("..Bedroom|..Bedrooms",real[i].find('ul', {'class':'aux-info'}).get_text().strip())[0][0]
            except (IndexError,TypeError,AttributeError) as e:
                d["bedrooms"]= None 
            try:
                d["bathrooms"]= re.findall("..Bathroom|..Bathrooms",real[i].find('ul', {'class':'aux-info'}).get_text().strip())[0][0]
            except (IndexError,TypeError,AttributeError) as e:
                d["bathrooms"]= None
            try:
                d["toilets"] = re.findall("..Toilet|..Toilets",real[i].find('ul', {'class':'aux-info'}).get_text().strip())[0][0]
            except (IndexError,TypeError,AttributeError) as e:
                d["toilets"] = None
            try:
                d["garage"] = re.findall("..Parking Space",real[i].find('ul', {'class':'aux-info'}).get_text().strip())[0][0]
            except (IndexError,TypeError,AttributeError) as e:
                d["garage"] = None
            try:
                d["location"] = real[i].find('address', {'class':'voffset-bottom-10'}).get_text().strip()
            except (IndexError,TypeError,AttributeError):
                d["location"] = None
            try:
                d["location_s"] = real[i].find('address', {'class':'voffset-bottom-10'}).get_text().strip().split()[-2]
            except(IndexError,TypeError,AttributeError):
                d['location_s'] = None
            try:
                d["price"]=real[i].find('span', {'class':'pull-sm-left'}).get_text().strip().split('₦')[1].split(' ')[0]
            except (IndexError,TypeError,AttributeError) as e:
                d["price"] = None            
            scraped.append(d)
            #print(l)
```

```python
df = pd.DataFrame(scraped)
df
```

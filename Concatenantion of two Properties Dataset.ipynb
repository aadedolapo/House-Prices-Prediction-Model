```python
def toilets(df1):
    if df1['bathrooms'] == 0:
        return 1
    if df1['bathrooms'] == 1:
        return 2
    if df1['bathrooms'] == 2:
        return 3
    if df1['bathrooms'] == 3:
        return 4
    if df1['bathrooms'] == 4:
        return 5
    if df1['bathrooms'] == 5:
        return 6
    if df1['bathrooms'] == 6:
        return 7
    if df1['bathrooms'] == 7:
        return 8
    else:
        return 9
```

```python
df1['toilets'] = df1.apply(toilets, axis=1)
```

```python
df1['garage'] = df1['toilets']
```

```python
df1.drop('toilets', axis=1,inplace=True)
```

```python
df1.rename(columns={'garage':'toilets'},inplace=True)
```
